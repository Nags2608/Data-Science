```python
# List of DataFrame names for the years 2015 to 2023
dfcys = [ "2015", "2016", "2017", "2018", "2019", "2020", "2021", "2022", "2023"]

# Extract the columns of 2015
dfc_2015_columns = dfc['2015'].columns

# Create modified datasets for the specified years with matched columns from 2007
dfc1 = {}

for year in dfcys:
    df = dfc[year].copy()  # Create a copy to avoid modifying in-place
    matched_columns = [col for col in df.columns if col in dfc_2015_columns]
    df = df[matched_columns]
    dfc1[year] = df  # Store the modified DataFrame in the dictionary

# Now, modified_datasets contains the modified versions of 2010 to 2022
# Access them using modified_datasets["2010"], modified_datasets["2011"], etc.
```
